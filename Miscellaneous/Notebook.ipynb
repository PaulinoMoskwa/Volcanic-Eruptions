### **1.3.** Download images from the url

```python
# Consider only events from 2015 as dataset
data = df_from_2015

# How many events are there?
len(data)
```

```python
import requests

# For every event:
for i in range(len(data)):

    # Url of the immagine
    url = images[data.iloc[i]['Volcano Name']]
    
    # Name of the image
    name = data.iloc[i]['Volcano Name']
    
    # Save the image (.jpg) with the name of the Volcano
    img_data = requests.get(url).content
    with open('images/' + name + '.jpg', 'wb') as handler:
        handler.write(img_data, )
```

### **1.4.** Resize the images

```python
from PIL import Image
import os

path = "images/"
dirs = os.listdir(path)

# Resize images (in-place) in size 400x230
def resize():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((400,230), Image.ANTIALIAS)
            imResize.save(f + '.jpg', 'JPEG', quality=90)

resize()
```

## **2.** Map

### **2.1.** Import libraries and define utilities

```python
# Libraries
from folium import IFrame
import folium
import base64

# Initial settings
resolution = 75
width      = 7
height     = 3
station    = '42'
lon, lat   = 15.004, 37.734
m          = folium.Map(location=[lat, lon], zoom_start=5)
```

```python
# Define a function to figure out whether a string is nan
def isnan(value):
    try:
        import math
        return math.isnan(float(value))
    except:
        return False
```

```python
# Define a function to enter each event on the map starting from:
#     - volcano name
#     - country where it is located
#     - year/month/day of eruption
#     - whether it caused a tsunami
#     - whether it caused an earthquake

def add_element_to_map(name, country, year, month, day, tsunami, earthquake):

   png = 'images/' + name + '.jpg'
   encoded = base64.b64encode(open(png, 'rb').read())

   # -----------------------------------------------------
   # TSUNAMI      ‚ùå
   # EARTHQUAKE   ‚ùå
   # -----------------------------------------------------
   if ((isnan(tsunami) and isnan(earthquake))):
      html = f'''
         <font face="Arial">
            <table>
               <tr>
                  <th>üóª</th>
                  <td>{name}</td>
               </tr>
               <tr>
                  <th>üåç</th>
                  <td>{country}</td>
               </tr>
               <tr>
                  <th>üí•</th>
                  <td>{year}</td>
            </table>
         </font><br>
      '''
      html    = html + '''<img src="data:image/png;base64,{}">'''
      html    = html.format
      iframe  = IFrame(html(encoded.decode('UTF-8')), width=(width*resolution)-110, height=(height*resolution)+120)
      popup   = folium.Popup(iframe, max_width=2650)
      icon    = folium.Icon(color="blue", icon="ok")
      marker  = folium.Marker(location=[lon, lat], popup=popup, icon=icon)
      marker.add_to(m)

   # -----------------------------------------------------
   # TSUNAMI      ‚ùå
   # EARTHQUAKE   ‚úÖ 
   # -----------------------------------------------------
   elif (isnan(tsunami)):
      html = f'''
         <font face="Arial">
            <table>
               <tr>
                  <th>üóª</th>
                  <td>{name}</td>
               </tr>
               <tr>
                  <th>üåç</th>
                  <td>{country}</td>
               </tr>
               <tr>
                  <th>üí•</th>
                  <td>{year}</td>
            </table>
         </font><br>
         <font face="Arial" color="#B53737">
            <b>In {year} the eruption generated an earthquake! ‚õîÔ∏è</b><br><br>
         </font>
      '''
      html    = html + '''<img src="data:image/png;base64,{}">'''
      html    = html.format
      iframe  = IFrame(html(encoded.decode('UTF-8')), width=(width*resolution)-110, height=(height*resolution)+170)
      popup   = folium.Popup(iframe, max_width=2650)
      icon    = folium.Icon(color="red", icon="ok")
      marker  = folium.Marker(location=[lon, lat], popup=popup, icon=icon)
      marker.add_to(m)

   # -----------------------------------------------------
   # TSUNAMI      ‚úÖ
   # EARTHQUAKE   ‚ùå
   # -----------------------------------------------------
   elif (isnan(earthquake)):
      html = f'''
         <font face="Arial">
            <table>
               <tr>
                  <th>üóª</th>
                  <td>{name}</td>
               </tr>
               <tr>
                  <th>üåç</th>
                  <td>{country}</td>
               </tr>
               <tr>
                  <th>üí•</th>
                  <td>{year}</td>
            </table>
         </font><br>
         <font face="Arial" color="#B53737">
            <b>In {year} the eruption generated a tsunami! ‚õîÔ∏è</b><br><br>
         </font>
      '''
      html    = html + '''<img src="data:image/png;base64,{}">'''
      html    = html.format
      iframe  = IFrame(html(encoded.decode('UTF-8')), width=(width*resolution)-110, height=(height*resolution)+170)
      popup   = folium.Popup(iframe, max_width=2650)
      icon    = folium.Icon(color="red", icon="ok")
      marker  = folium.Marker(location=[lon, lat], popup=popup, icon=icon)
      marker.add_to(m)

   # -----------------------------------------------------
   # TSUNAMI      ‚úÖ
   # EARTHQUAKE   ‚úÖ
   # -----------------------------------------------------
   else:
      html = f'''
         <font face="Arial">
            <table>
               <tr>
                  <th>üóª</th>
                  <td>{name}</td>
               </tr>
               <tr>
                  <th>üåç</th>
                  <td>{country}</td>
               </tr>
               <tr>
                  <th>üí•</th>
                  <td>{year}</td>
               </table>
         </font><br>
         <font face="Arial" color="#B53737">
            <b>In 2019 the eruption generated a tsunami and an earthquake! ‚õîÔ∏è</b><br><br>
         </font>
      '''
      html    = html + '''<img src="data:image/png;base64,{}">'''
      html    = html.format
      iframe  = IFrame(html(encoded.decode('UTF-8')), width=(width*resolution)-110, height=(height*resolution)+190)
      popup   = folium.Popup(iframe, max_width=2650)
      icon    = folium.Icon(color="red", icon="ok")
      marker  = folium.Marker(location=[lon, lat], popup=popup, icon=icon)
      marker.add_to(m)
```

### **2.2** Insert the events on the map

```python
for i in range(len(data)):
    lon        = float(data.iloc[i]['Coordinates'].split(',')[0])
    lat        = float(data.iloc[i]['Coordinates'].split(',')[1])
    name       = data.iloc[i]['Volcano Name']
    country    = data.iloc[i]['Country']
    tsunami    = data.iloc[i]['Flag Tsunami']
    earthquake = data.iloc[i]['Flag Earthquake']
    year       = data.iloc[i]['Year']
    month      = data.iloc[i]['Month']
    day        = data.iloc[i]['Day']
    add_element_to_map(name, country, year, month, day, tsunami, earthquake)
    print('Added volcano ', name, ', which erupted in ', year)
```
